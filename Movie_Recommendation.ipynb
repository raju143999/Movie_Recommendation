```python
def predict_movie(movie_name):
    movie_rating=movie_table[movie_name]
    corr_with_movie=movie_table.corrwith(movie_rating)
    movie_dataframe=pd.DataFrame(corr_with_movie,columns=["corelation"])
    movie_dataframe.dropna(inplace=True)
    cr_movie=movie_dataframe.join(ratings["num_ratings"])
    prediction=cr_movie[cr_movie['num_ratings']>100].sort_values("corelation",ascending=False)
    return prediction    
```
